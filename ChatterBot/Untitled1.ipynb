```python
import pandas as pd
from chatterbot import ChatBot
```

```python
tyrion_bot = ChatBot("Tyrion")
```

```python
tyrion_bot= ChatBot(
    'Tyrion',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite:///database.sqlite3'
)
```

```python
tyrion_bot = ChatBot(
    'tyrion_bot',  
    logic_adapters=[
        'chatterbot.logic.BestMatch'],
)
```

```python
from chatterbot.trainers import ListTrainer
trainer = ListTrainer(tyrion_bot)
```

```python
with open('listDialogues.txt', 'r') as filehandle:
    lines= filehandle.readlines()
 
```

```python
list_dial = [l.replace('\n','') for l in lines]
```
