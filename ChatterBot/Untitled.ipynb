```python
got_dialogues= got_data_initial.loc[:][['Name','Sentence']]
got_dialogues
```

```python
got_tyrion_added= got_dialogues.loc[got_dialogues.Name=='tyrion']['Sentence']
got_tyrion_added
```

```python
got_tyrion_added.index -1
```

```python
got_tyrion_context= got_dialogues.loc[got_tyrion_added.index -1]['Sentence']
got_tyrion_context
```

```python
got_dialogues= pd.concat([got_tyrion_added,got_tyrion_context])
got_dialogues=got_dialogues.sort_index()
got_dialogues
```
